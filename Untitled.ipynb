```python
# example spectrogram plotting -- redo to get the raw values

# spectrogram code for 30s audio samples
from matplotlib.ticker import FuncFormatter
import librosa

def audio_to_spectrogram(
    audio_path,
    target_sr=22050,
    duration=30.0,
    n_fft=1024,
    hop_length=256,
    n_mels=128,
    fmin=50,
    fmax=11025,
    db_range=80.0,
    plot=False,
    save_path=None,
    yscale="log" #pass in either log or linear
):
    """
    Convert a field audio clip to a normalized log-mel spectrogram.

    Parameters
    ----------
    audio_path : str
        Path to audio file (flac from Sage node)
    target_sr : int
        Resample to this rate. 22050 Hz captures up to 11 kHz — sufficient
        for most wildlife/environmental sounds.
    duration : float
        Clip length in seconds. Shorter clips are zero-padded; longer are
        truncated. Enforces a fixed output shape for batching.
    n_fft : int
        FFT window size. 1024 ≈ 46 ms at 22050 Hz — resolves transient
        sounds like bird calls (typically 50–200 ms).
    hop_length : int
        Samples between frames. 256 = ~11.6 ms step (75% overlap with
        n_fft=1024). Balances time resolution and array size.
    n_mels : int
        Number of mel filter banks. 128 is standard for audio ML (BirdNET,
        PANNs, AudioMAE). Compresses high-frequency bins where energy is
        sparse in natural soundscapes.
    fmin : float
        Lowest mel filter center (Hz). 50 Hz cuts wind/handling noise.
    fmax : float
        Highest mel filter center (Hz). Default is Nyquist of target_sr.
    db_range : float
        Dynamic range to retain (dB). Clips the noise floor so quiet
        background doesn't dominate the normalized output.
    plot : bool
        Display spectrogram inline.
    save_path : str or None
        If provided, save the plot to this path.
    yscale : str
        "log" or "linear" — controls the y-axis DISPLAY only, for
        visual inspection. Does not change the underlying `spec` array
        returned below; that's identical either way since mel binning
        already determines the actual frequency resolution.
        - "log": stretches low frequencies over more pixels, making
          periodic/rhythmic low-frequency structure (e.g. a fan cycling,
          a mechanical pulse) easier to spot visually.
        - "linear": easier to read in plain kHz (whole-number ticks),
          better for quickly orienting on where bird/insect activity
          sits (typically 1-8 kHz), but can visually compress low-
          frequency detail into a blur.
        Use "log" when debugging/exploring a clip, "linear" for a
        quick human-readable overview. Either way, the same `spec`
        array goes into the encoder — this only affects what you see.

    Returns
    -------
    spec : np.ndarray, shape (n_mels, time_steps), dtype float32
        Mel spectrogram normalized to [0, 1]. Treat as a single-channel
        image when feeding an embedding model.
    mel_freqs : np.ndarray
        Center frequency of each mel bin (Hz).
    times : np.ndarray
        Center time of each frame (s).
    """
    # load, resample to target_sr, mix to mono — librosa handles all three
    data, _ = librosa.load(audio_path, sr=target_sr, mono=True, duration=duration)

    # pad if clip is shorter than expected duration
    n_samples = int(duration * target_sr)
    if len(data) < n_samples:
        data = np.pad(data, (0, n_samples - len(data)))

    # mel spectrogram (power, not amplitude)
    S = librosa.feature.melspectrogram(
        y=data, sr=target_sr,
        n_fft=n_fft, hop_length=hop_length,
        n_mels=n_mels, fmin=fmin, fmax=fmax,
        window="hann",
    )

    # convert to dB relative to peak, clip dynamic range, normalize to [0, 1]
    S_db = librosa.power_to_db(S, ref=np.max, top_db=db_range)
    spec = (S_db + db_range) / db_range  # [-db_range, 0] → [0, 1]
    spec = spec.astype(np.float32)

    mel_freqs = librosa.mel_frequencies(n_mels=n_mels, fmin=fmin, fmax=fmax)
    times = librosa.frames_to_time(
        np.arange(spec.shape[1]), sr=target_sr, hop_length=hop_length
    )

    if plot or save_path:
        fig, ax = plt.subplots(figsize=(12, 4))

        if yscale == "linear":
            # display in kHz with whole-number ticks
            freqs_display = mel_freqs / 1000
            ylabel = "Frequency (kHz)"
        else:
            # display in raw Hz, log scale
            freqs_display = mel_freqs
            ylabel = "Frequency (Hz, mel scale)"

        img = ax.pcolormesh(times, freqs_display, spec, shading="gouraud", cmap="inferno")
        plt.colorbar(img, ax=ax, label="Normalized Power")

        if yscale == "log":
            ax.set_yscale("log")
        else:
            from matplotlib.ticker import FuncFormatter
            ax.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{int(x)}"))

        ax.set_ylabel(ylabel)
        ax.set_xlabel("Time (s)")
        ax.set_title(audio_path)
        plt.tight_layout()

        if save_path:
            plt.savefig(save_path, dpi=150, bbox_inches="tight")
        if plot:
            plt.show()
        plt.close(fig)

    return spec, mel_freqs, times


# example usage:
audio_file = "sample_audio.flac"
spec, freqs, times = audio_to_spectrogram(audio_file, plot=True, yscale="linear") #change the yscale for y axis and viz display

print(f"Output shape: {spec.shape}  →  ({len(freqs)} freq bins × {len(times)} time steps)")
print(f"Freq range: {freqs[0]:.0f} – {freqs[-1]:.0f} Hz")
print(f"Time range: {times[0]:.2f} – {times[-1]:.2f} s")
print(f"Value range: [{spec.min():.3f}, {spec.max():.3f}]")
```
